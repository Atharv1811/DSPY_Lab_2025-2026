1. Explain input and output characteristics for f_oneway() function from scipy.stats


Ans:- The f_oneway() function is used to perform one-way ANOVA.

Input:

It accepts two or more arrays (or lists/Series) of numerical values.

Each array should represent a group of samples.

Example from your notebook:

from scipy.stats import f_oneway

f_oneway(group1, group2, group3)

Each group might represent a different treatment, category, or experimental condition.

Output:

Returns two values:

F-statistic: The ratio of variance between the groups to the variance within the groups.

p-value: Tells whether the observed differences among means are statistically significant.

If p-value < 0.05, we reject the null hypothesis and conclude that at least one group mean is significantly different.

2. Explain results of sm.stats.anova_lm() function

Ans:- This function is used for more detailed ANOVA analysis using a linear model.

Input:

Takes a fitted OLS model as input (created using ols() from statsmodels.formula.api).

Example:

import statsmodels.api as sm

from statsmodels.formula.api import ols

model = ols('Score ~ C(TeachingMethod)', data=df).fit()

sm.stats.anova_lm(model)

Output:

Returns a table with the following columns:

Source	df	sum_sq	mean_sq	F	PR(>F)

TeachingMethod	2	400.25	200.13	6.28	0.003

Residual	30	956.3	31.88

df: Degrees of freedom for each source.

sum_sq: Sum of squares for each source.

mean_sq: Mean sum of squares (sum_sq / df).

F: F-statistic.

PR(>F): p-value (Probability that the F-statistic occurred by chance).

If PR(>F) is less than 0.05, we conclude that the factor has a significant effect on the dependent variable.

3. What role does ANOVA play in data analytics?

Ans:- ANOVA helps in:

Comparing multiple group means at once without performing many t-tests.

Identifying important factors affecting a continuous variable.

Reducing dimensionality by focusing on significant variables.

Supporting feature selection in machine learning pipelines.

In short, ANOVA helps you figure out which categorical variables significantly influence a numerical outcome.

4. When to use One-Way ANOVA? Explain with real-world case.

Ans:- Use one-way ANOVA when:

You have one categorical independent variable with 2 or more groups.

You want to see if it significantly affects a numerical dependent variable.

Example:

You want to know whether students' performance is influenced by the type of teaching method used:

Groups: Traditional, Online, Hybrid

Dependent variable: Final Exam Score

You would use one-way ANOVA to test:

Do exam scores differ significantly across the three teaching methods?

5. When to use Two-Way ANOVA? Explain with real-world case.

Ans:- Use two-way ANOVA when:

You have two categorical independent variables.

You want to see their individual effects and interaction effect on a numerical variable.

Example:

Suppose you're analyzing crop yield based on:

Fertilizer type (A, B, C)

Irrigation level (Low, Medium, High)

Here:

Two factors: Fertilizer and Irrigation

Dependent variable: Crop yield

Two-way ANOVA can tell you:

Does fertilizer type affect yield?

Does irrigation level affect yield?

Is there an interaction between fertilizer and irrigation?
