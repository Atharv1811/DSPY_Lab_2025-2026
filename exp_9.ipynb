1. When do we go for Chi-square test?

Ans:- The Chi-square test is used when we want to examine the relationship between two categorical variables. It helps answer questions like:

Is there an association between a person's education level and their job type?

Does gender influence purchase decisions?

In your Kaggle notebook, the variables Gender and Bought are both categorical, making them suitable for a Chi-square test.

2. Do we use Chi-square test for independence? Explain how.

Ans:- Yes, the Chi-square test is commonly used to test for independence between two categorical variables.

How it works:

A contingency table is created using the two variables.

The Chi-square test calculates the difference between the observed and expected frequencies in each cell.

If the difference is large enough (statistically significant), we conclude that the variables are not independent.

Example from your notebook:

from scipy.stats import chi2_contingency

table = pd.crosstab(data['Gender'], data['Bought'])

chi2, p, dof, expected = chi2_contingency(table)

If the p-value is less than 0.05, we reject the null hypothesis (that the variables are independent).

3. Explain the role of Critical value and p-value in Chi-square test.

Ans:- Critical Value: This is the cutoff point obtained from the Chi-square distribution table based on the desired significance level (e.g., 0.05) and degrees of freedom. If the test statistic exceeds the critical value, we reject the null hypothesis.

p-value: This represents the probability of obtaining the observed results, assuming the null hypothesis is true. If the p-value is less than the significance level (commonly 0.05), we reject the null hypothesis.

In most statistical software (like Python with chi2_contingency), the p-value is directly provided and is commonly used over the critical value approach.

4. Can we use Chi-square test to show existence of correlation between two features in a dataset? How will you interpret the result to indicate correlation presence?

Ans:- The Chi-square test does not measure correlation in the statistical sense. Instead, it detects whether there is an association or dependency between two categorical variables.

Interpretation of results:

If p-value < 0.05, conclude that there is a statistically significant association between the variables.

If p-value ≥ 0.05, conclude that there is no evidence of association between the variables.

To measure the strength of the association, you can calculate Cramér’s V, which gives a value between 0 (no association) and 1 (strong association).

5. What is degrees of freedom and how to calculate it while performing Chi-square test?

Ans:- Degrees of freedom (df) represent the number of independent values that can vary in the calculation. For a contingency table:

Formula:

df=(r−1)×(c−1)df = (r - 1) \times (c - 1)

Where:

r = number of rows in the table

c = number of columns

In your case, if Gender has 2 categories and Bought also has 2 categories:

df=(2−1)×(2−1)=1df = (2 - 1) \times (2 - 1) = 1
