Q.1) Describe all three data set in terms of following : how many features, their categories (num,text,categorical, nominal,ordinal etc), number of tuples, missing values.

Ans:- a) IRIS Dataset

Number of features: 4 input features + 1 target variable

Feature types:
Numerical: Sepal length, Sepal width, Petal length, Petal width
Categorical (Nominal): Species (Setosa, Versicolor, Virginica)

Number of tuples (rows): 150

Missing values: No missing values (clean dataset)

(b) Air Quality Dataset

Number of features: Typically ~13–15 features (depends on version)

Feature types:
Numerical: CO, NO2, O3, temperature, humidity, etc.
Date/Time: Timestamp features (can be treated as temporal data)
Some datasets may include categorical (e.g., station location)
Number of tuples: ~9000+ records (hourly measurements over time)
Missing values: Yes, contains many missing values (often represented as -200 or NaN)

(c) Wine Quality Dataset

Number of features: 11 input features + 1 output (quality)

Feature types:
Numerical: All physicochemical properties (pH, alcohol, acidity, etc.)
Output (quality): Ordinal categorical (ranges from 0–10, usually 3–8)
Number of tuples:
Red wine: 1599
White wine: 4898
Missing values: No missing values (well-structured dataset)


Q2) In IRIS data set which non graphical EDA is performed ?

Ans:- Non-graphical EDA includes statistical summary methods such as:

Mean, median, mode
Standard deviation, variance
Minimum and maximum values
Correlation matrix
In the IRIS dataset, summary statistics are performed to understand feature distribution and central tendency.

For example:

Petal length has a strong correlation with species
Setosa has distinctly smaller petal measurements compared to others
Thus, descriptive statistics and correlation analysis are the main non-graphical EDA techniques used.


Q3). In  Air Quality data set which graph is used to perform multivariate analysis?

Ans:- For multivariate analysis in the Air Quality dataset, commonly used graphs include:

Heatmap (Correlation Matrix Heatmap)
Shows relationships between multiple variables at once
Helps identify strong positive/negative correlations
Other possible graphs:

Pair plot (scatterplot matrix)
3D scatter plot
But the most commonly used and effective graph is the Heatmap, as it clearly visualizes correlations among many variables simultaneously.


Q4). In Wine quality data set can you do non graphical EDA ? Explain your answer.

Ans:- Yes, non-graphical EDA can be performed on the Wine Quality dataset.

Explanation:
Since all features are numerical, statistical methods can be easily applied:

Mean, median, variance for each chemical property
Correlation analysis (e.g., alcohol vs quality)
Frequency distribution of quality scores
Insights obtained:
Alcohol content positively correlates with quality
Volatile acidity negatively affects quality
Quality values are not evenly distributed (imbalanced dataset)
Thus, non-graphical EDA is not only possible but very effective for this dataset.
