1. What is the need for Hypothesis Testing?

Ans:- Hypothesis testing is used to:

Make decisions or judgments about a population based on sample data.

Check whether a claim or assumption (called a hypothesis) about the population is true.

Avoid conclusions based on random chance.

2. Steps in Hypothesis Testing (with diagram)

Ans:- Steps:

State the Hypotheses

Null Hypothesis (H₀)

Alternate Hypothesis (H₁)

Set the Level of Significance (α)

Common value: 0.05 (95% confidence)

Choose the Test Statistic

Use t-test or z-test depending on the data

Calculate the Test Statistic

Use formula or software

Make a Decision

Compare p-value with α or t-value with critical value

Draw Conclusion

If p < α, reject H₀

If p ≥ α, accept H₀

3. What is Student's t-distribution?

Ans:- A probability distribution used when the sample size is small (usually less than 30).

Used when population standard deviation is unknown.

It is similar to the normal distribution but has thicker tails.

4. What is the level of significance?

Ans:- It is the probability of rejecting the null hypothesis when it is actually true.

Common levels: 0.05 (5%), 0.01 (1%)

A smaller alpha means stricter decision criteria.

5. What role does t play in hypothesis testing?

Ans:- The t-value is the test statistic calculated from the sample.

It shows how far the sample mean is from the population mean, measured in terms of standard error.

It is compared to the critical t-value to decide whether to accept or reject H₀.

6. Hypothesis Testing Example: Is average age of TE IT students 19?

Ans:- Step 1: Define Hypotheses

H₀: μ = 19

H₁: μ ≠ 19

Step 2: Take a sample

Example:
ages = [18, 20, 19, 21, 19, 18, 20, 19, 19, 20]

Step 3: Use Python to perform t-test

from scipy import stats

ages = [18, 20, 19, 21, 19, 18, 20, 19, 19, 20]
pop_mean = 19

t_stat, p_val = stats.ttest_1samp(ages, pop_mean)

print(f"t-statistic = {t_stat:.4f}")
print(f"p-value = {p_val:.4f}")
Step 4: Decision

If p-value > 0.05 → Accept H₀

If p-value < 0.05 → Reject H₀

Output:

t-statistic = 0.8165
p-value = 0.4372
Since 0.4372 > 0.05, we accept H₀

Conclusion: The average age of TE IT students is approximately 19.
