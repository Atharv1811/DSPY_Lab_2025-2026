
1. What is a One-Tailed and Two-Tailed T-Test?


Ans:- Two-Tailed T-Test:

Tests whether the sample mean is significantly different from the population mean in either direction (greater or smaller).

Used when: You are only interested in whether it's different, not the direction.

Example:

H₀: μ = 19

H₁: μ ≠ 19

One-Tailed T-Test:

Tests if the sample mean is either significantly greater than or less than the population mean.

You choose one direction before testing.

Example (greater):

H₀: μ ≤ 19

H₁: μ > 19

2. What are Type I and Type II Errors?

Ans:-
Type I Error (False Positive):

Rejecting the null hypothesis when it is actually true.

Example:

You conclude that the average age is not 19, but in reality, it is.

Type II Error (False Negative):

Failing to reject the null hypothesis when it is actually false.

Example:

You accept that the average age is 19, but in reality, it is not.


3. What is the Power of a Statistical Test?

Ans:- Power = Probability of correctly rejecting a false null hypothesis

It is calculated as:

Power = 1 - Probability of Type II Error (β)

A high power means the test is more likely to detect a real effect if it exists.

4. Explain input and output of stats.ttest_rel

Ans:- Input:

Takes two related samples (e.g., before and after scores of same students)

stats.ttest_rel(sample1, sample2)

Both lists should have same number of elements

Output:

Returns:

t-statistic: The value calculated by the test

p-value: Probability of observing the result if H₀ is true

Use Case: When comparing measurements taken from the same group at two times.



5. Function to find Power of a Statistical Test in statsmodels

Ans:- In statsmodels.stats.power, the function is:

TTestPower().power(effect_size, nobs, alpha, alternative)

Output:

Returns a value between 0 and 1

Closer to 1 → High power → Test can detect true effect

Closer to 0 → Low power → Test might miss a real effect

Example:

from statsmodels.stats.power import TTestPower

test = TTestPower()

power = test.power(effect_size=0.5, nobs=30, alpha=0.05)

print(power)

If power = 0.85 → There’s an 85% chance the test will detect a true difference.
